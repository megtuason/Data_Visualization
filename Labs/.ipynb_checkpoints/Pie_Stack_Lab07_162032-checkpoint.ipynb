```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("pokemon2.csv")
df2 = pd.read_csv("outbreaks.csv")
```

```python
grass = df[df.type1=="grass"]
```

```python
grass = grass.groupby("type2")["name"].count()
```

```python
new = pd.DataFrame(grass)
new.name
```

```python
fig = plt.gcf()
population = new
segments = 'dark', 'fairy', 'fighting','flying', 'ghost', 'grass', 'ground', 'ice', 'poison', 'psychic', 'steel'
fig.set_size_inches(8,8)
pie = plt.pie(population, labels=segments, startangle=90, autopct = '%1.0d%%', rotatelabels = True)
for tx in pie[1]:
    rot = tx.get_rotation()
    tx.set_rotation(rot)
plt.show()
```

```python
CA = df2[df2.State=="California"]
OH = df2[df2.State =="Ohio"]
MA = df2[df2.State=="Massachusetts"]
```

```python
CA = CA[CA.Location=="Restaurant"]
OH = OH[OH.Location=="Restaurant"]
MA = MA[MA.Location=="Restaurant"]
```

```python
CA = pd.DataFrame( CA.groupby("Year")["Illnesses"].sum())
OH = pd.DataFrame( OH.groupby("Year")["Illnesses"].sum())
MA = pd.DataFrame( MA.groupby("Year")["Illnesses"].sum())
```

```python
states = ["California", "Ohio", "Massachusetts"]
plt.stackplot( CA.index, CA.Illnesses, OH.Illnesses, MA.Illnesses, labels=states)
plt.legend(loc= "upper right")
plt.title("Illnesses Contracted from Restaurants")
plt.xlim(1998,2015)
plt.show()
```
