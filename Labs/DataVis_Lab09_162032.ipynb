```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
heart = pd.read_csv("heart.csv")
```

```python
from numpy import median
```

```python
sns.violinplot(x = 'cp', y = "chol", data = heart, hue = 'sex', split = True )
```

1. A violin plot is almost like a box plot wherein it is able to show the medians and ranges of values more effectively. This plot also has sort of like a box plot in the middle to show values of the confidence interval and the median. In addition to this, the sides are like sideways density plots whose width represents higher or lower frequency.

2. In this case, the data shows that there is generally a higher probability for women to have chest pains at a given serum cholesterol level in these specific types of chest pains as compared to men. The median value of the data for all types is generally around 250mg of serum cholesterol. This implies a dangerous level of cholesterol that is more likely to cause chest pains among men and women. All this data shows that there is a higher probability for women to be at risk of a heart problem given the circumstances of the type of chest pains used in the data and the cholesterol levels.
