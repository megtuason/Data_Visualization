```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.options.mode.chained_assignment = None

%matplotlib inline
```

```python
data = pd.read_csv('admis.csv')
data
```

```python
topuniv = data[data["UnivRating"] >= 3]
tu = topuniv[topuniv["ChanceOfAdmission"] > 0.70]

tu = topuniv.groupby("UnivRating")["GRE"].mean()
```

```python
meani = tu.describe()["mean"]
mean = [meani]*3
```

```python
tu = pd.DataFrame(tu)
```

```python
x = tu.GRE
y = tu.index

plt.title("Average GRE Scores Per University Rating")
plt.barh(y,x)
plt.plot(mean, y, ls = "--", label = "Average GRE", color = "red")

plt.text(mean[0],y[0], str("%.2f" %meani), color = "red")

labels = tu.GRE
count = 0
for i in range(3,6):
    plt.text(x[i], y[count], str("%.2f" %labels[i]), horizontalalignment = "left", verticalalignment = "center")
    count += 1

plt.yticks([3,4,5])
plt.xlabel("GRE Score")
plt.ylabel("University Rating")
plt.xlim(300,350)

plt.legend()
plt.show()
```
